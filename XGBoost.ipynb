### 4) Split the data into training and testing sets.

```python
train = df.loc[df.index < '01-01-2015']
test = df.loc[df.index > '01-01-2015' ]
```

### 7) Creating Time-based Features for Dataset.

```python
def create_features(df):
    df = df.copy()
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.day_of_week
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.day_of_year
    return df

df = create_features(df)
```

```python
df.head()
```

### 8) Visualize our Feature Dataset by Hour and Month.

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df, x="hour", y = "PJME_MW")
ax.set_title("MW by Hour")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df, x="month", y = "PJME_MW", palette= "Blues")
ax.set_title("MW by Month")
plt.show()
```

```python
train = create_features(train)
test = create_features(test)
```

```python
FEATURES = ['hour', 'dayofweek', 'quarter', 'month', 'year', 'dayofyear']
TARGET = 'PJME_MW'
```

```python
X_train = train[FEATURES]
y_train = train[TARGET]

X_test = test[FEATURES]
y_test = test[TARGET]
```

### 9) Preparing Dataset and Training XGBoost Model

```python
import xgboost as xgb
reg = xgb.XGBRegressor(n_estimators = 1000,early_stopping_rounds=50,learning_rate = 0.01)
reg.fit(X_train,y_train,eval_set=[(X_train,y_train),(X_test,y_test)])
```

### 10) Feature Importance Analysis and Predictions using XGBoost Model.

```python
reg.feature_importances_
```

```python
fi = pd.DataFrame(data=reg.feature_importances_,
                 index = reg.feature_names_in_,
                 columns = ['importances'])
```

```python
fi
```

```python
fi.sort_values('importances')
```

```python
fi.sort_values('importances').plot(kind='barh', title="Feature Importance")
plt.show()
```

### 11) Merging Predictions into Original Dataset.

```python
test['prediction'] = reg.predict(X_test)
plt.plot(test['PJME_MW'],label='True')
plt.plot(test['prediction'],label='Prediction')
plt.legend()
```

```python
test['predictions'] = reg.predict(X_test)
df = df.merge(test['predictions'], how='left', left_index=True, right_index=True)
ax = df['PJME_MW'].plot(figsize=(15,5))
df['predictions'].plot(ax=ax, style='.')
plt.legend(["True Data", "Predictions"])
ax.set_title("Raw Data and Predictions")
plt.show()
```

```python
ax = df.loc[(df.index > "04-01-2018") & (df.index < "04-08-2018")]['PJME_MW'].plot(figsize=(15,5),title = "Week of Data");
df.loc[(df.index > '04-01-2018') & (df.index < '04-08-2018')]['predictions'].plot(style='.')
plt.legend(["True Data", "Predictions"])
plt.show()
```

### 12) Calculate the Errors and Find the Best and Worst Predictions.

```python
from sklearn.metrics import mean_squared_error
score = np.sqrt(mean_squared_error(test['PJME_MW'], test['predictions']))
print(f"RMSE Score on Test set: {score:0.2f}")
```
